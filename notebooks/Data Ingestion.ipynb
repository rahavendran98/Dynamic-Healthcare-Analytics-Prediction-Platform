```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
import os
import warnings
warnings.filterwarnings("ignore")
```

```python
DATA_DIR = r"D:\projects\Healthcare\data\raw\\"
```

```python
datasets = {
    "diabetes" : {"file":   "diabetes_prediction_dataset.csv",  "target": "diabetes", "size": 20000},
    "cardio"   : {"file":    "cardio_data_processed.csv",       "target": "cardio",   "size": 20000},
    "stroke"   : {"file":    "stroke_prediction.csv",           "target": "Diagnosis", "size":15000},
}

print("Setup ready. Datasets:", list(datasets.keys()))
```

```python
def slice_dataset(name, info):
    """Stratified method"""
    
   
    path = os.path.join(DATA_DIR, info["file"])
    
    
    df = pd.read_csv(path)
    target = info["target"]
    target_size = info["size"]
    
    print(f"\n{'='*55}")
    print(f"  {name.upper()}")
    print(f"{'='*55}")
    print(f"Original shape : {df.shape[0]} rows x {df.shape[1]} cols")
    
    # 3. Original target ratio 
    print(f"\nOriginal target ratio ('{target}'):")
    print((df[target].value_counts(normalize=True) * 100).round(2))
    
    # 4. 
    if df.shape[0] <= target_size:
        print(f"\n--> Already {df.shape[0]} rows (<= {target_size}). As-is vechukrom.")
        sliced = df
    else:
        # 5. 
        sliced, _ = train_test_split(
            df,
            train_size=target_size,
            stratify=df[target],
            random_state=42
        )
        print(f"\n--> Sliced to {target_size} rows (stratified).")
    
    # 6. Sliced ratio 
    print(f"\nSliced target ratio ('{target}'):")
    print((sliced[target].value_counts(normalize=True) * 100).round(2))
    
    return sliced
```

```python
sliced_datasets = {}

for name, info in datasets.items():
    sliced = slice_dataset(name, info)
    
    
    out_path = os.path.join(DATA_DIR, f"{name}_sliced.csv")
    sliced.to_csv(out_path, index=False)
    
    sliced_datasets[name] = sliced
    print(f"\nSaved --> {out_path}")
```
